```python
import tensorflow as tf 
import keras

from keras.utils import plot_model, to_categorical
from keras.models import Sequential, load_model
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.preprocessing.image import ImageDataGenerator

import cv2 as cv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split

import pickle


%matplotlib inline
sess = tf.compat.v1.Session(config=tf.compat.v1.ConfigProto(log_device_placement=True))
print(sess)
```

```python
pre_x = pickle.load(open(r"D:\CS7\Machine Learning Models\artx.txt", "rb"))
pre_y = pickle.load(open(r"D:\CS7\Machine Learning Models\arty.txt", "rb"))
```

```python
pre_x.shape
```

```python
counts = np.unique(pre_y, return_counts=True)
counts
```

```python
nulos = 0
minimo = 0
maximo = 5000
total = 0
cant = 0
for i in pre_x: 
    if i is None: 
        nulos += 1
        continue
    temp = i.shape[0] * i.shape[1]
    if temp > maximo:
        maximo = temp
    if temp < minimo:
        minimo = temp 
    total += temp 
    cant += 1
print(nulos)
print(minimo)
print(maximo)
print(total / cant)
```

```python
categories = ["Drawing","Engraving","Iconography","Painting","Sculpture"]

ax = sns.barplot(x=categories, y=counts[1], palette="Blues_d")
plt.title("Distribución por Categoría")
sns.despine(left=True,bottom=True)
plt.show()
```

```python
def clean_size(x,y,size):
    el = []
    for i in range(len(x)):
        if x[i] is None:
            el.append(i)
            continue
        else:
            x[i] = cv.resize(x[i],(size,size))
            x[i] = cv.cvtColor(x[i],cv.COLOR_BGR2RGB)
    x = np.array(list(np.delete(x,el)))
    y = np.delete(y,el)
    print("New shape of Dataset: ",x.shape)
    return (x,y)
    
sig_data = clean_size(pre_x,pre_y,64)
sig_x = sig_data[0]
sig_y = sig_data[1]
```

```python
# hacer la union de 2 categorias 
new_y = sig_y.copy()
for i in range(len(sig_y)):
    if sig_y[i] == 0:
        new_y[i] = 1
    new_y[i] -= 1

counts_new = np.unique(new_y, return_counts=True)
counts_new
```

```python
categories_new = ["Drawing/Engraving","Iconography","Painting","Sculpture"]

ax = sns.barplot(x=categories_new, y=counts_new[1], palette="Blues_d")
plt.title("Distribución por Categoría")
sns.despine(left=True,bottom=True)
plt.show()
```

```python
new_y.shape
```

```python
#final_y = to_categorical(new_y)
final_y = new_y.copy()
final_x = sig_x.copy()
```

```python
print(final_y.shape)
print(final_x.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(final_x,final_y,test_size=0.2,random_state=2)
x_train.shape
```

```python
no_usar = load_model('./fourth.h5')
```

```python
main1 = Sequential()

main1.add(Conv2D(128, (3,3), activation='relu', input_shape=(x_train.shape[1:])))
main1.add(MaxPooling2D(pool_size=(2,2)))
main1.add(Dropout(0.25))

main1.add(Conv2D(64, (3,3), activation='relu'))
main1.add(Conv2D(64, (3,3), activation='relu'))
main1.add(MaxPooling2D(pool_size=(2,2)))
main1.add(Dropout(0.25))

main1.add(Conv2D(32, (3,3), activation='relu'))
main1.add(Conv2D(32, (3,3), activation='relu'))
main1.add(MaxPooling2D(pool_size=(2,2)))
main1.add(Dropout(0.25))

main1.add(Flatten())
main1.add(Dense(512,activation='relu'))
main1.add(Dense(256,activation='relu'))
main1.add(Dense(4,activation='softmax'))

main1.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

print(main1.summary())
```

```python
main1_train = main1.fit(x_train,y_train,
                    batch_size=64,
                    epochs=25,
                    validation_split=0.2,
                    shuffle=True,
                    verbose=1)
```

```python
score, acc = main1.evaluate(x_test, y_test,
                            batch_size=64)
print('Test score:', score)
print('Test accuracy:', acc)

plt.plot(main1_train.history['accuracy'], label='accuracy')
plt.plot(main1_train.history['val_accuracy'], label = 'val_accuracy')
plt.plot(main1_train.history['loss'], label = 'loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.2, 1])
plt.legend(loc='lower left')
```

```python
main1.save("./clean/main1.h5")
pickle.dump(main1_train,open(r"./clean/main1.txt", "wb"))
```

```python
main2 = Sequential()

main2.add(Conv2D(32, (3,3), activation='relu', input_shape=(x_train.shape[1:])))
main2.add(MaxPooling2D(pool_size=(2,2)))
main2.add(Dropout(0.25))

main2.add(Conv2D(64, (3,3), activation='relu'))
main2.add(Conv2D(64, (3,3), activation='relu'))
main2.add(MaxPooling2D(pool_size=(2,2)))
main2.add(Dropout(0.25))

main2.add(Conv2D(128, (3,3), activation='relu'))
main2.add(Conv2D(128, (3,3), activation='relu'))
main2.add(MaxPooling2D(pool_size=(2,2)))
main2.add(Dropout(0.25))

main2.add(Flatten())
main2.add(Dense(512,activation='relu'))
main2.add(Dropout(0.3))
main2.add(Dense(256,activation='relu'))
main2.add(Dense(4,activation='softmax'))

main2.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

print(main2.summary())
```

```python
main2_train = main2.fit(x_train,y_train,
                    batch_size=64,
                    epochs=30,
                    validation_split=0.2,
                    shuffle=True,
                    verbose=1)
```

```python
score, acc = main2.evaluate(x_test, y_test,
                            batch_size=64)
print('Test Loss:', score)
print('Test accuracy:', acc)


plt.plot(main2_train.history['accuracy'], label='Train Accuracy', color="blue")
plt.plot(main2_train.history['val_accuracy'], label = 'Train Validation Accuracy')
plt.plot(main2_train.history['loss'], label = 'Loss', color="purple")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.2, 1])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2))
plt.title("Historia del Entrenamiento")
plt.box(False)
plt.savefig("./history.png")
plt.show()
```

```python
main2.save("./clean/main2.h5")
pickle.dump(main2_train,open(r"./clean/main2.txt", "wb"))
```

```python
def plot_features(result):
    number = result.shape[3]
    cant = number / 4
    fig = plt.figure(figsize=(20,20))
    for i in range(number):
        axtemp = fig.add_subplot(cant,4,i+1)
        fig.tight_layout(pad=0.0)
        plt.imshow(result[0,:,:,i],cmap='gray')
        plt.axis("off")
    plt.show()
plot_features(result_first_conv)
```

```python
def plot_interesting(arreglo,results):
    for i in arreglo:
        plt.imshow(results[0,:,:,i],cmap='gray')
        plt.axis("off")
        plt.show()

interesantes = [3,13,25,29]
plot_interesting(interesantes,result_first_conv)
```
